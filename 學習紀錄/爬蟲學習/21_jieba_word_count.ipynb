```python
import jieba
import os
#載入mydict字典 使分割的字可以合併為字典的字並統計
jieba.load_userdict("./mydict.txt")
```

```python
flist = os.listdir("./PttGossiping")
flist
```

```python
sb= ""
for textfile in flist:
    with open("./PttGossiping/%s"%(textfile), "r",encoding="utf-8") as f:
        sb += f.read().split("---split---")[0]
sb
```

```python
s_cut = jieba.cut(sb)
```

```python
for w in s_cut:
    print(w)
```

```python
for w in s_cut:
    print(w)
```

```python
s_cut = jieba.cut(sb)
wc = {}   #自訂空字典

for w in s_cut:    #從段詞內抓
    if len(w) < 2:  #長度小於2
        continue    #略過
    if not w in wc:  #如果不在裡面,為1
        wc[w] = 1
    else:    #否則,+1
        wc[w] += 1
wc
```

```python
wc_list = [(w,wc[w]) for w in wc]
wc_list            
```

```python
wc_list=[]
for w in wc:
    tmpdata = tuple([w, wc[w]])
    wc_list.append(tmpdata)
wc_list
```

```python
#做排序 沒有reverse=True為小到大  如有 則 大到小
wc_list.sort(key=lambda t: t[1],reverse=True)
wc_list
```
